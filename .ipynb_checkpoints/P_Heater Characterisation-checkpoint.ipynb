### F1 -> laser in O1 powermeter in O3

```python
# want zero phase to correspond to signal going straight through filter, not to herald. So zero phase at min here
mytunics.laser_wavelength(s_wavelength)
time.sleep(2)
guess = [0.10,3,1e-3,1e-3]
phase_offset = 1*np.pi

background_voltage =0
sleep=0.5
voltages, currents, alpha, beta, gamma, powers_mW, c, phi_0, A, B = h.characterise_heater(heaters, powermeter1, 'F1', {}, v_max=5.9, sleep=sleep, background_voltage=background_voltage, phase_offset=phase_offset, fringe_fit_guess=guess,steps = 30)
print(f'fit visibility = {A/B}')
print(f'data visibility = {(powers_mW.max() - powers_mW.min())/(powers_mW.max() + powers_mW.min())}')
```
