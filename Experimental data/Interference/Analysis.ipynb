```python
from qutip import *
import pandas as pd
import numpy as np
import scipy

import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
df = pd.read_csv("drift_over_time_1584448946.csv")
df
```

```python
df['# #time'] = (df['# #time'] - df['# #time'][0])
```

```python
Y_results = [df['output_1'], df['output_2']]
```

```python
fig, ax1 = plt.subplots(figsize=(15,8))
colours = ['tab:green', 'tab:red', 'tab:blue']

for index, y_result in enumerate(Y_results):
    x = df['# #time']
    y = savgol_filter(y_result,11,2)
    lns1 = ax1.plot(x, y, color=colours[index], markersize=7, linewidth=3)
    
ax1.set_xlabel('Time (s)', fontsize=20)
ax1.set_ylabel('Power (dB/m)', fontsize=20)
ax1.set_xlim(0,1100)
ax1.set_ylim(-43,-35)
ax1.tick_params(labelsize=20)
# plt.plot([0, 0], [0, 1.7], 'r--', lw=2)
# plt.grid()

ax1.legend(['Ch1','Ch2',], fontsize=15, loc=4)

plt.savefig('Classic_Phase_Drift', bbox_inches='tight')

plt.show()
```
